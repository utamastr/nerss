```python
def calc_LT(list_TLE, t):
    df = pd.DataFrame([], columns=['LT'])
    for i in range(0, len(list_TLE)):
        # print(i)
        if list_TLE[i][0]=='1':
            L_ = list_TLE[i].split(' ')
            epoch_year = '20' + L_[5][0:2]
            epoch_day = L_[5][2:]
            epoch = ts.utc(float(epoch_year), 1, float(epoch_day), 0, 0, 0)
            if epoch - t[0]>0:
                if epoch - t[-1]<0:
                    L1 = list_TLE[i-2]
                    L2 = list_TLE[i-1]
                    sat =  EarthSatellite(L1, L2, 'LA2', ts)
                    t_now = t[t-sat.epoch>0]
                    t_now = t_now[t_now-epoch<0]
                    # print(t_now.tt.size)
                    if t_now.tt.size != 0:
                        geocentric = sat.at(t_now)
                        r = wgs84.geographic_position_of(geocentric)
                        R_GCRS = geocentric.position.km
                        x = R_GCRS[0,]
                        y = R_GCRS[1,]
                        z = R_GCRS[2,]
                        lat = r.latitude.degrees
                        lon = r.longitude.degrees
                        alt = r.elevation.km
                        ra_sat = np.arctan2(y,x)*np.rad2deg(1)/15
                        ra_sun, de_sun, d_sun = (earth).at(t_now).observe(sun).apparent().radec()
                        ra_sun_h = ra_sun.hours
                        LT_sat = ra_sat - ra_sun_h + 12
                        d = {'lon' : lon, 'lat' : lat, 'LT' : LT_sat}
                        df = pd.concat([df, pd.DataFrame(data=d)], ignore_index=True)

                    # print(len(t_now), len(x))
                else:
                    break
    return df
```

```python
def coverage(df_sat, max_lat, duration):
    t_min = duration[0]
    t_max = duration[1]
    lat = df_sat['lat']
    lon = df_sat['lon']
    LT = df_sat['LT']
    
    # filter time
    lon = lon[LT > t_min]
    lat = lat[LT > t_min]
    LT = LT[LT > t_min]
    lon = lon[LT < t_max]
    lat = lat[LT < t_max]
    LT = LT[LT < t_max]
    
    # filter lat
    lon = lon[lat > - max_lat]
    LT = LT[lat > - max_lat]
    lat = lat[lat > - max_lat]
    lon = lon[lat < max_lat]
    LT = LT[lat < max_lat]
    lat = lat[lat < max_lat]
    
    d = 0.01
    xgrid = np.arange(-180, 180, d)
    ygrid = np.arange(-max_lat, max_lat, d)
    h, xedges, yedges = np.histogram2d(lon, lat, bins=(xgrid, ygrid))
    
    d = {'lat' : lat, 'lon' : lon, 'LT' : LT}
    df_out = pd.DataFrame(data=d)
    return h, xedges, yedges
    
h, xe, ye = coverage(df_A3, 11, [9.44, 9.73])
print(h.sum())
h, xe, ye = coverage(df_A2, 11, [9.44, 9.73])
print(h.sum())
# fig, ax = plt.subplots(1,1, figsize=(18,10))
# ax.pcolormesh(xe, ye, h.T)
# ax.axis('equal')
```
